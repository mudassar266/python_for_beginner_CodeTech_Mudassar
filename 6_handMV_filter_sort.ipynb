```python
# handling missing values by filling with mean
df_original = pd.read_csv('births.csv')
# fill missing values with mean
df_original['day'].fillna(df_original['day'].median(), inplace=True)
```

## Step.2 Data Filtering 

```python
# filter kren yr 2000 k bad waly bachy 
df_2000 = df_original[df_original['year'] >= 2000]
df_2000
```

```python
# kis month men sab sy zyada bachy paida huy hen saal 2001 mn
df_2001 = df_original[df_original['year'] == 2001]
df_2001
```

## Step.3 sorting 

```python
df_sorted = df_original.sort_values(by='births', ascending=False)
df_sorted
```

```python
# sorting data by year and month
df_sorted_yr_mnth = df_original.sort_values(by=['year', 'month'], ascending=[True, False])
df_sorted_yr_mnth
```

```python
# line plot
plt.figure(figsize=(10, 6))
df_original.groupby('year')['births'].sum().plot(marker = 'o', color = 'b')
plt.title('Total births per year')
plt.xlabel('Year')
plt.ylabel('Total births')  
plt.grid()
plt.show()
```

```python
# bar plot
plt.figure(figsize=(10, 6))
df_original.groupby('year')['births'].sum().plot(kind='bar', color = 'r')
plt.title('Total births per year')
plt.xlabel('year')
plt.ylabel('Total births')
plt.grid()
plt.show()
```

```python
# bar chart of births by gender
plt.figure(figsize=(10, 6))
df_original.groupby('gender')['births'].sum().plot(kind='bar', color = ['pink', 'blue'])
plt.xlabel('gender')
plt.ylabel('Total births')
plt.title('Total births by gender')
plt.show()
```
